```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read File
df = pd.read_csv("student.csv")
# Show First five rows
print(df.head())
# Show Total number of rows and column in data
print(df.shape)
```

```python
# Show Average of marks
print(df["marks"].mean())
# Filter by high marks
print(df[df["marks"] > 80])
# Prints Maximum marks
print(df["marks"].max())
```

```python
# Filter highscorers to save into another file 
high_scorers = df[df["marks"] > 80]
# Create and save  highscorers data to file 
high_scorers.to_csv("high_scorers.csv",index=False)
# Print Succesfully
print("High scorers saved succesfully ")
print(high_scorers)
```

```python
# Bar chart of student marks
plt.figure(figsize=(8, 4))
plt.bar(df['name'], df['marks'], color='skyblue')
plt.title("Student Marks Bar Chart")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("marks_chart.png") 
plt.show()
```
